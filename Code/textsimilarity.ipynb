```python
import jieba.posseg as pseg
import codecs
import warnings
import os,sys
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
from gensim import corpora, models, similarities
import pandas as pd
import csv
```

```python
stop_words = 'G:/数据挖掘大作业/yd停用词.txt'
stopwords = codecs.open(stop_words,'r',encoding='utf8').readlines()
stopwords = [ w.strip() for w in stopwords ]
```

```python
#结巴分词后的停用词性 [标点符号、连词、助词、副词、介词、时语素、‘的’、数词、方位词、代词]
stop_flag = ['x', 'c', 'u','d', 'p', 't', 'uj', 'm', 'f', 'r']
```

```python
def fre(words):
    word_dict= {} 
    for item in words:
        if item not in word_dict: 
            word_dict[item] = 1  
        else:  
            word_dict[item] += 1  
    return word_dict
```

```python
def tokenization(filename):
    result = []
    with open(filename, 'r',encoding='utf8') as f:
        text = f.read()
        words = pseg.cut(text)
    for word, flag in words:
        if flag not in stop_flag and word not in stopwords:
            result.append(word)
    wordDic = fre(result)
    result = [item for item in result if wordDic[item]>3]
    return result
```

```python
#选取几篇待检测文章。
file_dir = "G:\\数据挖掘大作业\\东野圭吾作品全集TXT\\东野圭吾全集\\training\\"
file_list = os.listdir(file_dir)
filenames = []
for file in file_list:
    filenames.append(file_dir+file)
corpus = []
for each in filenames:
    print(each)
    try:
        corpus.append(tokenization(each))
    except UnicodeDecodeError as e:
        pass
print(len(corpus))
```

```python
dictionary = corpora.Dictionary(corpus)
with open('G:/数据挖掘大作业/mid_dic','a+') as f:
    for key,value in dictionary.iteritems():
        print(key,value)
```

```python
doc_vectors = [dictionary.doc2bow(text) for text in corpus]
print(len(doc_vectors))
print(doc_vectors)

#建立TF-IDF模型
tfidf = models.TfidfModel(doc_vectors)
tfidf_vectors = tfidf[doc_vectors]
print(len(tfidf_vectors))
print(len(tfidf_vectors[0]))

```

```python
file_dir2 = "G:\\数据挖掘大作业\\东野圭吾作品全集TXT\\东野圭吾全集\\testing\\"
file_list2 = os.listdir(file_dir2)
filenames2 = []
for file in file_list2:
    filenames2.append(file_dir+file)
with open('G:/数据挖掘大作业/输出文档/相似度分析/千禧年前后TF-IDF相似度.csv','w+',encoding = 'utf8') as csvf:
    writer = csv.writer(csvf)        
    for i in range(len(filenames2)):
        query = tokenization(file_dir2+file_list2[i])
        query_bow = dictionary.doc2bow(query)
# print(len(query_bow))
# print(query_bow)
        index = similarities.MatrixSimilarity(tfidf_vectors)
#用TF-IDF模型计算相似度
        tfidf_query_bow = tfidf[query_bow]
        sims = index[tfidf_query_bow]
        writer.writerow(list(enumerate(sims)))
        print(list(enumerate(sims)))
```
